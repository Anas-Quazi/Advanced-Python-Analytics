```python

#^ read csv file

file_path = r'C:\Users\anasq\Downloads\Wearable_Health_Lifestyle_Ethics.csv'
df = pd.read_csv(file_path)
print('dataset uploaded successfully!')
```

```python
print('first 5 records \n')
print(df.head())

print('Shape : ')
print(df.shape)

print('Colm Info')
print(df.info())

print('statistical summary \n')
print(df.describe())

df['Gender'] = df['Gender'].replace('Non-binary', 'Male')
```

```python
print('missing values/colm : ')
print(df.isnull().sum())
```

```python

#~ visualization

#todo numeric distribution 
numeric_cols = df.select_dtypes(include=np.number).columns

# List of colors (ensure it's at least as long as numeric_cols)
colors = ['skyblue', 'salmon', 'lightgreen', 'orange', 'purple']

for i, col in enumerate(numeric_cols):
    # Select color by index (using % to loop back if cols > colors)
    df[col].hist(color=colors[i % len(colors)]) 
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python

#* correlational analysis
print('correlational matrix')
correlational_matrix = df.corr(numeric_only=True)
print(correlational_matrix)

plt.figure()
plt.imshow(correlational_matrix)
plt.colorbar()
plt.title('Correlation heatmap')
plt.xticks(range(len(correlational_matrix.columns)), correlational_matrix.columns)
plt.yticks(range(len(correlational_matrix.columns)), correlational_matrix.columns)
plt.show()
```

```python

#^ step 6 : ethical analysis section

#? privacy risk detection
sensitive_keywords = ['name', 'email', 'phone', 'address',
                      'location', 'gps', 'id', 'dob']
privacy_risk = []

for col in df.columns:
    for keyword in sensitive_keywords:

        #* spelling mistake
        if keyword.lower() in col.lower():
            privacy_risk.append(col)

if privacy_risk:
    print('privacy risk detected!')
    print('sensitive column found : ', privacy_risk)
else:
    print('no sensitive keyword')

#^ bias detection
demographic_cols = ['Gender', 'Age', 'Region']

for col in demographic_cols:
    if col in df.columns:
        print(f'Distribution of {col} : ')
        print(df[col].value_counts(normalize=True))

print('Ethical concern : ')
print('If one demographic group dominate dataset')
```
